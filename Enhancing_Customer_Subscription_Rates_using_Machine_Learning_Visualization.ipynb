Exploratory Data Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the aesthetic style of the plots
sns.set_style("white")  # Simple and clean background

# Create a figure with two subplots (axes) side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# First scatter plot with 'Gender' as the hue
sns.scatterplot(ax=axes[0], data=data, x='Unit price', y='Rating', hue='Gender')
axes[0].set_title('Unit Price vs Rating by Gender')
axes[0].set_xlabel('Unit Price')
axes[0].set_ylabel('Rating')

# Second scatter plot with 'Customer type' as the hue
sns.scatterplot(ax=axes[1], data=data, x='Unit price', y='Rating', hue='Customer type')
axes[1].set_title('Unit Price vs Rating by Customer Type')
axes[1].set_xlabel('Unit Price')
axes[1].set_ylabel('Rating')

# Display the plot
plt.tight_layout()  # Adjusts the subplots to give some spacing between them
plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Setting up the figure
plt.figure(figsize=(14, 10))
sns.set_style('whitegrid')

# KDE for 'Unit price'
plt.subplot(2, 3, 1)
sns.kdeplot(data['Unit price'], fill=True)
plt.title('Density Plot for Unit Price')

# KDE for 'Quantity'
plt.subplot(2, 3, 2)
sns.kdeplot(data['Quantity'], fill=True)
plt.title('Density Plot for Quantity')

# KDE for 'Total'
plt.subplot(2, 3, 3)
sns.kdeplot(data['Total'], fill=True)
plt.title('Density Plot for Total')

# KDE for 'cogs'
plt.subplot(2, 3, 4)
sns.kdeplot(data['cogs'], fill=True)
plt.title('Density Plot for COGS')

# KDE for 'Rating'
plt.subplot(2, 3, 5)
sns.kdeplot(data['Rating'], fill=True)
plt.title('Density Plot for Rating')

# KDE for 'gross income'
plt.subplot(2, 3, 6)
sns.kdeplot(data['gross income'], fill=True)
plt.title('Density Plot for Gross Income')

plt.tight_layout()  # Adjusts subplots to give some spacing between them
plt.show()
```

```python
plt.style.use("default")  # Use the default style
plt.figure(figsize=(8,6))  # Set the figure size

# Create a line plot for the data slice
sns.lineplot(x="Rating", y="Unit price", data=data[170:180], marker='o')

# Adding title and labels
plt.title("Rating vs Unit Price", fontsize=15)
plt.xlabel("Rating")
plt.ylabel("Unit Price")

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use("default")
plt.figure(figsize=(8, 6))

sns.pointplot(x="Rating", y="Quantity", data=data[170:180], color='green', markers='x')

plt.title("Rating vs Quantity", fontsize=15)
plt.xlabel("Rating")
plt.ylabel("Quantity")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for column in data.columns:
    if data[column].dtype == 'object' or data[column].dtype.name == 'category':
        data[column] = le.fit_transform(data[column])

# Print the DataFrame to see the transformed categorical columns
print(data)
```
